```python
import urllib.request
import zipfile
import os
import pandas as pd

# 1. Download do Dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00228/smsspamcollection.zip"
zip_path = "../data/raw/smsspamcollection.zip"
extract_path = "../data/raw/"

# Baixa e extrai se não existir
if not os.path.exists(extract_path + "SMSSpamCollection"):
    print("Baixando o dataset...")
    urllib.request.urlretrieve(url, zip_path)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print("Download e extração concluídos!")

# 2. Carregamento e Leitura
# O arquivo original é separado por tabulações (\t) e não tem cabeçalho
df = pd.read_csv(
    extract_path + "SMSSpamCollection", 
    sep='\t', 
    names=["label", "mensagem"]
)

# 3. Primeiras Estatísticas
print("Tamanho do Dataset:", df.shape)
print("\nDistribuição das Classes (Spam vs Ham):")
print(df['label'].value_counts())

display(df.head())
```
